```python
import json
import numpy as np


def save_config(path, instruments, params, strategy_params, params_for_fit=None):
    config = {
        "instruments_list": instruments,
    }
    config.update(params)
    config.update(strategy_params)
    if params_for_fit is not None:
        config,update(params_for_fit)
    with open(path, 'w') as f:
        json.dump(config, f)
```
